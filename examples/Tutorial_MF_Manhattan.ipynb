```python
import os
import sys
import math
path_base = os.getcwd()+"/.."
if path_base not in sys.path:
    sys.path.append(path_base)

import numpy as np
import matplotlib
import matplotlib.animation
import matplotlib.patches
import matplotlib.pyplot as plt


import shapely
import mfjet 
```

# Minkowski functionals and Valuations in Manhattan Geometry

In this tutorial, we will compute Minkowski functionals of the following points dilated by a square.

```python
# coordinates of points
coords = np.array([
    [-1,-1],
    [-1,1],
    [1,-1],
])
# dilation scales
arr_r = np.linspace(0,5,101)
```

```python
# visualizing points
fig = plt.figure(figsize=[3,3])
plt.plot(
    *(coords.T),
    "o"
)
plt.gca().set_xlim([-2,2])
plt.gca().set_ylim([-2,2])
plt.gca().set_xlabel("x", fontsize=15)
plt.gca().set_ylabel("y", fontsize=15)
```

```python
# initialize calculator
calculator = mfjet.MFManhattanCalculator()

# evaluate minkowski functionals given points and dilation scales.
arr_mfs = calculator.calc_mfs(coords, arr_r)
print(arr_mfs.shape)  # [:,0]: Euler char., [:,1]: boundary length, [:,2]: area
```

# Visualizing change of Minkowski functionals over dilation scale

If topological and geometrical fetures are not changing over dilation, the changing of Minkowski functionals over dilation scale respects the following Steiner-type formula.
  * $\mathrm{MF}_0(r) = \mathrm{MF}_0$
  * $\mathrm{MF}_1(r) = \mathrm{MF}_0 \times C_{01} r + \mathrm{MF}_1 \times C_{11}$
  * $\mathrm{MF}_2(r) = \mathrm{MF}_0 \times C_{02} r^2 + \mathrm{MF}_1 \times C_{21} r + \mathrm{MF}_2 \times C_{21}$

In the limit of $r \rightarrow 0$, the dilation is not changing geometric structure of three dots. The Steiner-type formula uses Euler characteristic 3, boundary length 0, and area 0.
  * $\mathrm{MF}_0(r) \rightarrow 3$
  * $\mathrm{MF}_1(r) \rightarrow 3 \times 8 r + 0$
  * $\mathrm{MF}_2(r) \rightarrow 3 \times 4 r^2 + 0 \times r + 0$

In the limit of $r \rightarrow \infty$, the dilation is seeing the three dots as L-shaped line. The Steiner-type formula uses Euler characteristic 1, boundary length 4, and area 0.
  * $\mathrm{MF}_0(r) \rightarrow 1$
  * $\mathrm{MF}_1(r) \rightarrow 1 \times 8 r + 4 \times 2$
  * $\mathrm{MF}_2(r) \rightarrow 1 \times 4 r^2 + 4 \times 2 r + 0$

```python
fig = plt.figure(figsize=[7,7]);

# dilation figure
#####################################
plt.subplot(2,2,1)
plt.plot(
    coords[:,0],
    coords[:,1],
    "o",
    c="black"
)

list_rects = [
    matplotlib.patches.Rectangle(
        coord-0.5, 
        width=0.5*2, 
        height=0.5*2, 
        fill=True, 
        lw=None, 
        fc="turquoise"
    )
    for coord in coords
]
for rect in list_rects:
    plt.gca().add_patch(rect)


line1, = plt.plot(
    [], [],
    "--",
    color="black",
    lw=0.5
)
line2, = plt.plot(
    [], [],
    "--",
    color="black",
    lw=0.5
)

plt.gca().set_xlabel("x",fontsize=15)
plt.gca().set_ylabel("y",fontsize=15)
plt.gca().set_xlim([-5,5])
plt.gca().set_ylim([-5,5])

# MF0 = Euler characteristic
#####################################
plt.subplot(2,2,2)
marker_euler, = plt.plot(
    [], [],
    "o",
    c="Blue"
)
plt.plot(
    arr_r,
    arr_mfs[:,0],
    c="red",
    label=r"Euler characteristic"
)
plt.plot(
    arr_r,
    3 * np.ones_like(arr_r),
    "--",
    c="gray",
    label=r"asymptote as $r \rightarrow 0$"
)
plt.plot(
    arr_r,
    1 * np.ones_like(arr_r),
    ":",
    c="gray",
    label=r"asymptote as $r \rightarrow \infty$"
)
plt.legend()
plt.gca().set_xlabel("r",fontsize=15)
plt.gca().set_ylabel("Euler Characteristic",fontsize=15)
plt.gca().set_xlim([0,5])
plt.gca().set_ylim([-2,5])

# MF1 = Boundary length
#####################################
plt.subplot(2,2,3)
marker_length, = plt.plot(
    [], [],
    "o",
    c="Blue"
)
plt.plot(
    arr_r,
    arr_mfs[:,1],
    c="red",
    label=r"Boundary length"
)
plt.plot(
    arr_r,
    3 * 8 * arr_r,
    "--",
    c="gray",
    label=r"asymptote as $r \rightarrow 0$"
)
plt.plot(
    arr_r,
    8 * arr_r + 8,
    ":",
    c="gray",
    label=r"asymptote as $r \rightarrow \infty$"
)
plt.legend()
plt.gca().set_xlabel("r",fontsize=15)
plt.gca().set_ylabel("Boundary length",fontsize=15)
plt.gca().set_xlim([0,5])
plt.gca().set_ylim([0,50])

# MF2 = area
#####################################
plt.subplot(2,2,4)
marker_area, = plt.plot(
    [], [],
    "o",
    c="Blue"
)
plt.plot(
    arr_r,
    arr_mfs[:,2],
    c="red",
    label=r"Area"
)
plt.plot(
    arr_r,
    3 * 4 * arr_r**2,
    "--",
    c="gray",
    label=r"asymptote as $r \rightarrow 0$"
)
plt.plot(
    arr_r,
    4 * arr_r**2 + 8 * arr_r,
    ":",
    c="gray",
    label=r"asymptote as $r \rightarrow \infty$"
)
plt.legend()
plt.gca().set_xlabel("r",fontsize=15)
plt.gca().set_ylabel("Area",fontsize=15)
plt.gca().set_xlim([0,5])
plt.gca().set_ylim([0,100])

plt.tight_layout()

def animate(i):
    r = arr_r[i]
    marker_euler.set_data([r],[arr_mfs[i,0]])
    marker_length.set_data([r],[arr_mfs[i,1]])
    marker_area.set_data([r],[arr_mfs[i,2]])
    for rect in list_rects:
        xy = rect.get_xy()
        center = xy + np.array([rect.get_width(), rect.get_height()])/2.
        rect.set_xy(center-r)
        rect.set_width(2*r)
        rect.set_height(2*r)
        
    # phase change at r=1
    if r < 1.:
        line1.set_data([], [])
        line2.set_data([], [])
    else:
        line1.set_data([-1,-1], [-1,1])
        line2.set_data([-1,1], [-1,-1])

animate(15)
plt.show()

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=arr_r.shape[0]);
```
